```python
import pandas as pd
import torch
import ast
import numpy as np
import os

# 1. Load the original CSV
csv_path = "/home/sunil/am2/poetry-demo/AM2_Poly_biodegradebilty/DNN_Models/Data_Cleaned_M0L-DNN_Model/Mol_DNN_cleaned.csv"
save_path = "/home/sunil/am2/poetry-demo/AM2_Poly_biodegradebilty/DNN_Models/Data_Cleaned_M0L-DNN_Model/MOL_DNN_processed.pt"
df = pd.read_csv(csv_path)

print(f"📖 Loaded {len(df)} rows from CSV.")
```

```python
# 3. Convert 'fingerprints' column from strings to a float tensor
# This is the step that was missing. It parses "[0.1, 0.2...]" into a real list.
print("🔄 Converting fingerprints... (this may take about 30-60 seconds)")
fps_list = df['fingerprints'].apply(ast.literal_eval).tolist()
fps_tensor = torch.tensor(fps_list, dtype=torch.float32)

# 4. Extract Time and Target
# We ensure these are float32 for PyTorch compatibility
time_tensor = torch.tensor(df['time_day'].values, dtype=torch.float32)
target_tensor = torch.tensor(df['biodegradation'].values, dtype=torch.float32)
```

## PyTorch Tensor Conversion and Export

```python
# 5. Save everything in a clean bundle
data_bundle = {
    'fingerprints': fps_tensor,
    'time_day': time_tensor,
    'biodegradation': target_tensor
}

# Ensure directory exists
os.makedirs(os.path.dirname(save_path), exist_ok=True)

# Save the bundle
torch.save(data_bundle, save_path)

print(f"✅ Success! Corrected bundle saved to: {save_path}")
print(f"📊 Final Check:")
print(f"   - Fingerprints Shape: {fps_tensor.shape}")  # Should be [2199, 600]
print(f"   - Time Shape: {time_tensor.shape}")          # Should be [2199]
print(f"   - Target Shape: {target_tensor.shape}")      # Should be [2199]
```
